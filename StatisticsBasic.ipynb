1. How do you calculate the mean, median, and mode of a dataset?

    -> Mean (Average)

    Formula: Mean = Sum of all values/ Number of values

    Example: For data: 2, 4, 6
    
    Mean = 2 + 4 + 6/ 3 = 4

    Median -
    
    Steps:

    Arrange values in order.

    If odd number of values → middle one is median.

    If even number → average of two middle values.

    Example: for data: 3, 1 , 4
    ordered: 1, 3, 4
    median = 3

    for data: 5, 2, 8, 6
    ordered: 2, 5, 6, 8
    median = 5 + 6/ 2 = 11/ 2= 5.5

    Mode (Most Frequent Value)
    Definition: Value that appears most often.

    Example: Data: 7, 7, 2, 3, 2, 2 → mode = 2 If no value repeats → no mode
    If multiple values repeat the same number of times → dataset is multimodal.
          

```python
#2. Write a Python program to compute the variance and standard deviation of a dataset.

data = [12, 15, 21, 20, 18, 17, 16]

#function to calculate mean

def calculate_mean(data):
    mean = sum(data) / len(data)
    print(f"Mean: {mean}")
    return mean

#function to calculate variance

def calculate_variance(data):
    mean = calculate_mean(data)
    variance = sum((x - mean) ** 2 for x in data) / len(data)
    print(f"Variance: {variance}")
    return variance


#function to calculate standard deviation

def calculate_std_dev(data):
    variance = calculate_variance(data)
    std_dev = variance ** 0.5
    print(f"Standard Deviation: {std_dev}")
    return std_dev
```

```python
#5. Write a Python function to calculate the range of a dataset

def calculate_range(data):
    if not data:
        print("Dataset is empty.")
        return None
    range_value = max(data) - min(data)
    print(f"Range: {range_value}")
    return range_value
```
